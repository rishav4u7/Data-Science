```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# import warning
import warnings
warnings.filterwarnings("ignore")
# pd.set_option("display.max_rows",None)
```

```python
df=pd.read_csv("Metrail+dataset.csv")
df.head()
```

```python
df.shape
```

```python
df=df.rename(columns={"Travellers('000)":"Travellers"})
df.columns
```

```python
df.info()
```

```python
df.plot(figsize=(12,4))
plt.show()
```

```python
# Forward fill
df = df.assign(Travellers_Forwardfill_Imputation=df.Travellers.fillna(method='pad'))
df['Travellers_Forwardfill_Imputation'].plot(figsize=(12, 4))
plt.legend(loc='best')
plt.title('Travellers passenger traffic: Forward imputation')
plt.show(block=False)
```

```python
df.mean()
```

```python
# background fill
df = df.assign(Travellers_backwardfill_Imputation=df.Travellers.fillna(method='bfill'))
df['Travellers_backwardfill_Imputation'].plot(figsize=(12, 4))
plt.legend(loc='best')
plt.title('Travellers passenger traffic: backward imputation')
plt.show(block=False)
```

```python
df.mean()
```

```python
# Linear Imputation fill
df = df.assign(Travellers_Linear_Interpolation_method_Imputation=df.Travellers.interpolate(method='linear'))
df['Travellers_Linear_Interpolation_method_Imputation'].plot(figsize=(12, 4))
plt.legend(loc='best')
plt.title('Travellers passenger traffic: LI imputation')
plt.show(block=False)
```
