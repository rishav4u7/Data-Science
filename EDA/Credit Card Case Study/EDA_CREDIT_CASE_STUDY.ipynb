- Columns with more Number of Unique data will fall in continuous variables

```python
#Checking outliers in AMT_ANNUITY
plt.figure(figsize=(12,2))
sns.boxplot(app['AMT_ANNUITY'])
```

```python
#calculating mean, median and mode for AMT_ANNUITY
print("Mean:",app['AMT_ANNUITY'].mean())
print("Median:",app['AMT_ANNUITY'].median())
print("Mode:",app['AMT_ANNUITY'].mode())
```

Inferences:
- There is a high range of outliers at left side of the distribution and hence imputing with mean would not be the right approach so imputing with median.
- Since median < mean, it is positvely skewed .That's why Imputing this column with median will help in the analysis

```python
app['AMT_ANNUITY'].fillna(app['AMT_ANNUITY'].median(),inplace=True)
```

```python
#Handling outlier in AMT_ANNUITY by excluding values outside 99%ile.
app=app[app.AMT_ANNUITY<np.nanpercentile(app['AMT_ANNUITY'], 99)]
#Rechecking outlier 
plt.figure(figsize=(12,2))
sns.boxplot(app['AMT_ANNUITY'])
```

```python
#Checking outliers in AMT_GOODS_PRICE
plt.figure(figsize=(12,2))
sns.boxplot(app['AMT_GOODS_PRICE']) 
```

```python
#calculating mean, median and mode for AMT_ANNUITY
print("Mean:",app['AMT_GOODS_PRICE'].mean())
print("Median:",app['AMT_GOODS_PRICE'].median())
print("Mode:",app['AMT_GOODS_PRICE'].mode())
```

```python
#Since both Median and mode are same and might lead to incorrect analysis . Hence imputing it with mean.
app['AMT_GOODS_PRICE'].fillna(app['AMT_GOODS_PRICE'].mean(),inplace=True)
```

```python
#Handling outlier in AMT_GOODS_PRICE by excluding values outside 99%ile.
app=app[app.AMT_ANNUITY<np.nanpercentile(app['AMT_GOODS_PRICE'], 99)]
#Rechecking outlier 
plt.figure(figsize=(12,2))
sns.boxplot(app['AMT_GOODS_PRICE'])
```

```python
#Checking outliers in AMT_INCOME_TOTAL
plt.figure(figsize=(12,2))
sns.boxplot(app['AMT_INCOME_TOTAL']);
```

```python
app.AMT_INCOME_TOTAL.quantile(0.99)
```

```python
app[app.AMT_INCOME_TOTAL>app.AMT_INCOME_TOTAL.quantile(0.99)].sort_values(by='AMT_INCOME_TOTAL', ascending=False).head(10)
```

```python
#There is a lot of difference between the 99th precentile and maximum value. Hence detecting Outlier by the IQR formula
iqr=app.AMT_INCOME_TOTAL.quantile(0.75)-app.AMT_INCOME_TOTAL.quantile(0.25)
AMT_INCOME_TOTAL_OUTLIER=app.AMT_INCOME_TOTAL.quantile(0.75)+(iqr*1.5)
AMT_INCOME_TOTAL_OUTLIER
```

```python
print(len(app[app.AMT_INCOME_TOTAL>AMT_INCOME_TOTAL_OUTLIER]))
print(len(app[app.AMT_INCOME_TOTAL>app.AMT_INCOME_TOTAL.quantile(0.99)]))
```

Inferences
- Since AMT_INCOME_TOTAL outlier is 337500 and 99th percentile is 450000 and the number of rows with these values are 12866 and 2750 respectively. Replacing the values higher than 450000.0 with 450000.0+10000.0 to make it continuous and distinctly identifiable and not 337500.0 since 99th percentile value is higher.


```python
app.loc[app['AMT_INCOME_TOTAL'] > 450000.0, ['AMT_INCOME_TOTAL']] = 450000.0+10000.0
```

```python
plt.figure(figsize = (10,2))
sns.boxplot(app.AMT_INCOME_TOTAL,color='b')
plt.show()
```

### Standarlizing dataset

```python
app.dtypes
```

```python
#CNT_FAM_MEMBERS cannot be float. Converting to integer
app.CNT_FAM_MEMBERS = app.CNT_FAM_MEMBERS.apply(lambda x: int(x))
```

```python
app.CNT_FAM_MEMBERS.dtype
```

```python
#DAYS_EMPLOYED, DAY_REGISTRATION, DAYS_ID_PUBLISH should be a positive value. Converting into absolute value
app.DAYS_EMPLOYED = app.DAYS_EMPLOYED.apply(lambda x: abs(x))
app.DAYS_REGISTRATION = app.DAYS_REGISTRATION.apply(lambda x: abs(x))
app.DAYS_ID_PUBLISH = app.DAYS_ID_PUBLISH.apply(lambda x: abs(x))
```

```python
#DAYS_BIRTH column is age of the peron at the time of loan application.
#This could be converted to age in years by dividing by 365.25(Considering leap years). Also it is with a 
#negative sign, hence needs to be treated.
app['Age'] = app.DAYS_BIRTH//(-365.25)
app.drop(['DAYS_BIRTH'],axis=1,inplace=True)
app.head()
```

```python
app.Age.describe()
```

```python
#Min age is 20 and max age is 68- Dividing into 7 intervals
app['Age_group'] = pd.cut(app['Age'],bins=np.linspace(10,70,7))
```

```python
app['Age_group']=app['Age_group'].astype(str)
app['Age_group'].value_counts()
```

```python
#Binning AMT_INCOME_TOTAL into income range categories for ease of analysis
app['IncomeRange'] = pd.qcut(app['AMT_INCOME_TOTAL'], q=[0,0.25,0.50,0.90,1], labels=['Low','Average','High','Very High'])
```

## Univariate Analysis
1. Categorical variables

```python
def plotfunc(plotv):
    plt.figure(figsize=(14,3))
    plt.subplot(1,2,1)
    sns.countplot(plotv,data=app_T0)
    plt.title('Distribution of '+ '%s' %plotv + ' for Non-Defaulters',fontsize=11,color='green')

    plt.xlabel(plotv,fontsize=10)
    plt.xticks(rotation=60)
    plt.ylabel('Number of cases for Non-Defaulters',fontsize=10)   
    
    plt.subplot(1,2,2)
    sns.countplot(plotv,data=app_T1)
    plt.title('Distribution of '+ '%s' %plotv + ' for Defaulters',fontsize=11,color='green')

    plt.xlabel(plotv,fontsize=10)
    plt.xticks(rotation=60)
    plt.ylabel('Number of cases for Defaulters',fontsize=10)
```

```python
plotfunc('NAME_CONTRACT_TYPE')
# We can notice that revolving loans are lesser in the defaulted population. Hence we can infer that
#revolving loans are comparatively safer. 
#There are around 250000 people has applied for cash loan who are not defaulters while more than 20000 people 
#in defaulters list have applied for loan.
```

```python
plotfunc('CODE_GENDER')

#Mostly females has aplied for loan whether they are from default category or non-default category. 
```
